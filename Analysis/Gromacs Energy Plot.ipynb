```python

import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the .xvg file, skipping lines starting with '#' and '@'
data = pd.read_csv('potential.xvg', comment='@', delim_whitespace=True, header=None, skiprows=17)
# Note: The value 17 in `skiprows` is based on the given file format. Adjust as needed.

# Split the data into time (ps) and potential (kJ/mol)
time = data[0]
potential = data[1]

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(time, potential, label="Potential", color="blue")
plt.xlabel("Time (ps)")
plt.ylabel("Potential (kJ/mol)")
plt.title("GROMACS Energies")
plt.legend()
plt.grid(True)
plt.tight_layout()

# Display the plot in the notebook
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the .xvg file, skipping lines starting with '#' and '@'
data = pd.read_csv('rmsd.xvg', comment='@', delim_whitespace=True, header=None, skiprows=17)
# Note: The value 17 in `skiprows` is based on the given file format. Adjust as needed.

# Split the data into time (ps) and potential (kJ/mol)
time = data[0]
potential = data[1]

data_2 = pd.read_csv('rmsd_xtal.xvg', comment='@', delim_whitespace=True, header=None, skiprows=17)
time_2 = data_2[0]
potential_2 = data_2[1]

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(time, potential, label="Ref:Equalibrated", color="black")
plt.plot(data_2[0], data_2[1], label="Ref:Crystal", color="red")
plt.xlabel("Time (ps)")
plt.ylabel("RMSD (nm)")
plt.title("RMSD")
plt.legend()
plt.grid(True)
plt.tight_layout()

# Display the plot in the notebook
plt.show()
```
