```python
import mdtraj as md
import numpy as np
import matplotlib.pyplot as plt

# Load the trajectory with the correct topology
traj = md.load_xtc('protein_only.xtc', top='protein_only.gro')

# Compute the distance between the first and last atom
def get_dist(traj):
    first = traj.xyz[:, 0, :]
    last = traj.xyz[:, -1, :]
    dist = np.sqrt(np.sum((first - last)**2, axis=1))
    return dist

dist = get_dist(traj)

# Compute the radius of gyration at each time
def get_gyra(traj):
    gyration = md.compute_rg(traj)
    return gyration

gyra = get_gyra(traj)

# Plotting the end-to-end distance
plt.figure(figsize=(10, 6))
plt.plot(traj.time, dist)
plt.xlabel('Time (ps)')
plt.ylabel('End-to-End Distance (nm)')
plt.title('End-to-End Distance vs. Time')
plt.grid(True)
plt.show()  # Display the figure

# Plotting the radius of gyration
plt.figure(figsize=(10, 6))
plt.plot(traj.time, gyra)
plt.xlabel('Time (ps)')
plt.ylabel('Radius of Gyration (nm)')
plt.title('Radius of Gyration vs. Time')
plt.grid(True)
plt.show()  # Display the figure
```
