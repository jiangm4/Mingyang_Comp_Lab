```python
import numpy as np
import matplotlib.pyplot as plt

# Load data
colvar_data = np.loadtxt('../Outputs/monitor/MONITOR-COLVAR.metadynLog')

# Extract time and reaction coordinate
time = colvar_data[:, 0] / 10  # Adjust if necessary
reaction_coord = colvar_data[:, 1]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(time, reaction_coord)
plt.title('Reaction Coordinate vs. Time')
plt.xlabel('Time (ns)')
plt.ylabel('Reaction Coordinate')  # Simplified label
plt.savefig('../Figures/reaction_coordinate_plot.png')
plt.show()
```
