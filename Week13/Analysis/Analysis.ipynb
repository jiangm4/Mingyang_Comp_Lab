```python
import numpy as np
import matplotlib.pyplot as plt

# Load data
nvt_data = np.loadtxt('../Outputs/NVT-1.ener')
npt_data = np.loadtxt('../Outputs/NPT-1.ener')

# Plot data
plt.figure(figsize=(10, 6))
plt.title('Potential Energy vs. Time')
plt.xlabel('Time (fs)')
plt.ylabel('Energy (a.u.)')

# Plot NVT data
plt.plot(nvt_data[:,1], nvt_data[:,4], label='NVT', color='blue')

# Plot NPT data
plt.plot(npt_data[:,1], npt_data[:,4], label='NPT', color='red')

# Add legend
plt.legend()

# Save the figure
plt.savefig('../Figures/potential_energy_plot.png')

# Show the plot
plt.show()
```
